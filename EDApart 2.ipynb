```python
import pandas as pd
import numpy as np

# ----- 1. Fix numeric columns -----
numeric_cols = ['Product_Category_1', 'Product_Category_2', 'Product_Category_3', 'Purchase']

for col in numeric_cols:
    # Ensure numeric type (coerce invalids to NaN)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Fill NaN with 0
    df[col] = df[col].fillna(0)

# ----- 2. Fix Stay_In_Current_City_Years -----
# Convert to string first to safely remove '+', then to numeric
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(str).str.replace('+','')
df['Stay_In_Current_City_Years'] = pd.to_numeric(df['Stay_In_Current_City_Years'], errors='coerce')
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].fillna(0)

# ----- 3. Convert Age to numeric -----
age_mapping = {
    '0-17': 0,
    '18-25': 1,
    '26-35': 2,
    '36-45': 3,
    '46-50': 4,
    '51-55': 5,
    '55+': 6
}

df['Age'] = df['Age'].map(age_mapping)
# Fill unmapped/NaN Age values with mode (most common age group)
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])

# ----- 4. Verify no NaNs remain -----
print(df.isna().sum())
```

```python
# 1. Fix numeric columns safely
numeric_cols = ['Product_Category_1', 'Product_Category_2', 'Product_Category_3', 'Purchase']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # invalid parsing → NaN
    df[col] = df[col].fillna(df[col].median())         # fill NaN with median

# 2. Fix Stay_In_Current_City_Years
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(str).str.replace('+','', regex=False)
df['Stay_In_Current_City_Years'] = pd.to_numeric(df['Stay_In_Current_City_Years'], errors='coerce')
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].fillna(0)

# 3. Map Age safely
age_mapping = {'0-17':0,'18-25':1,'26-35':2,'36-45':3,'46-50':4,'51-55':5,'55+':6}
df['Age'] = df['Age'].map(age_mapping)

# Use mode if available, otherwise default to -1
if df['Age'].mode().empty:
    df['Age'] = df['Age'].fillna(-1)  # -1 means unknown age
else:
    df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
# 4. Verify
print(df.isna().sum())
```
